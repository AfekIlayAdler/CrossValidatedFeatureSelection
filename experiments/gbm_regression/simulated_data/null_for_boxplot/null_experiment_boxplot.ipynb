```python
import sys
sys.path.append(r"C:\Users\afeki\Desktop\code\CrossValidatedFeatureSelection\experiments")  # path contains python_file.py
from plot_utils import get_regular_paths, get_data, plot_boxplot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
BOXPLOT = True
BARPLOT = False
metrics = ['ntrees', 'nleaves', 'error']
paths = get_regular_paths(one_hot=False)
dfs = []
for model, model_path in paths.items():
    fi, metrics = get_data(model, model_path)
    fi['algorithm'] = model
    dfs.append(fi)
    for temp_fi in fi.fi_type.unique():
        plot_boxplot(fi, title = F'{model}_{temp_fi}', save = True)
        
```

```python
all_data = pd.concat(dfs)
plt.figure(figsize=(30,10))
sns.boxplot(x="index", y="fi_value", hue = 'algorithm', data = all_data[all_data.fi_type =='permutation_test'] )
```
