```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from ast import literal_eval
```

```python
def get_results(exp, path):
    cols = [ 'gain']
    df = pd.read_csv(path, converters = {col: literal_eval for col in cols})
    results = pd.DataFrame()
    for col in cols:
        results[col] = pd.Series(df.loc[0,col])
    return results

paths = {
    'kfold_height' :"kfold_height.csv",
    'kfold_no_hyper' :"kfold_no_hyper.csv",
    "vanilla_hight" :"vanilla_height.csv",   
}
```

```python
for k, v in paths.items():
    try: 
        results = get_results(k,v)
        results = results.sort_index()
        
#         results = results[['gain']]
        results.plot(kind = 'bar', figsize = (10,6), title = k, ylim = (0,0.6))
        plt.plot()
    except:
        pass
```
